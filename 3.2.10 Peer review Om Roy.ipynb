```python
import pandas as pd
import numpy as np

products=pd.read_csv("products.csv")
products
```

```python
# 'Fire'

products[products['Description']\
         .apply(lambda x:'Fire' in x.title())]

# Lambda function is more simple thus more pythonic,
# used title() function as 'Fire' has a capital 'F'
#9 products contain word 'Fire' in description
```

```python
#'candle'

products[products['Description']\
.apply(lambda x:'candle' in x.lower())]

# Lambda function is more simple thus more pythonic,
# used lower() function as 'candle' is in lower case
# 178 products contain word 'candle' in description
```

```python
#'matches'
products[products['Description']\
.apply(lambda x:'matches'in x.lower())]

# Lambda function is more simple thus more pythonic,
# used lower() function as 'matches' is in lower case
# 6 products contain word 'matches' in description
```
